##### Lowercase and Punctation
Here we need to standardize our text by removing all the stopwords and thereby setting all characters to a lowercase, since our sole aim is to see how important each word is based on how frequently it was being usedn in the text

```python
## convertiung all characters to a lowercase
text_low = [w.lower() for w in tokenized_word]
print(text_low)
```

```python
## Next we remove tokens that are not alphabetic
text_rmv_punct = [word for word in text_low if word.isalpha()]
print(text_rmv_punct)
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(15,8))
f_dist = FreqDist(text_rmv_punct)
f_dist.plot(title='Frequency distribution of words', cumulative = True)
plt.show()
```

From the graph, we have the most appeared word displaying first. which are __'the'__, __'to'__, __'and'__ etc. These words do not really add meaning to the text,so they are considered as stopwords.

#### Stop words
These are those words that do not contribute to the deeper meaning of phrase. Therefore we can create a separate list of stopwords so it can be filtered out from our tokens

```python
#import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
print('to' in stop_words)
```

```python
# filtering stopwords from our tokens
clean_text = [w for w in text_rmv_punct if not w in stop_words]
print(clean_text)
```

Now that the stopwords are removed from our tokens, we can see which word appeared most often by rechecking our cumulative graph

```python
fig,ax = plt.subplots(figsize=(15,8))
f_dist = FreqDist(clean_text)
f_dist.plot(title='Frequency distribution of words', cumulative = True)
plt.show()
```

#### Regular Expression
This is a programming language in a special text string used for describing a search pattern. It is extremely useful for extracting information from text such as code, files,log, spredsheets or even documents.
Now lets see how can apply regex to match cedrtain patterns from our token

```python
import re
```

```python
# matching strings that ends with "ly"
[w for w in clean_text if re.search('ly$', w)]
```

```python
# matching strings that starts with "a" and ends with "y"
```

```python
[w for w in clean_text if re.search('^a.*y$', w)]
```

### Stem words
While working with words, we come across a lot of variations due to grammatical reasons. The concept of variations here means that we have to deal with different forms of words like democracy, democratic, and democratization.It is very neccessary for machines to understand that these different words have the same base form. In this way, it would be useful to extract the base forms while analyzing the text. This can be achieved by stemming in the heuritic prrocess by chopping off the ends of words.
Some applications, like documents classification, may benefit from stemming in order to both reduce the vocabulary and to focus on the the sense or sentiment of a document rather than deeper meaning.

```python
# stemming of words
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in clean_text]
```
